#Pandas Data Structure:-
1.Series:- It is One Dimensional Array With Indexes, It Stores Single Column or Row Of Data in Dataframe.

2.Dataframe:- It is a Tabular Spreadsheet Like Structure Representing Rows Each of Which Contains One or More Columns.

A One Dimensional Array Capable of Holding Any Tye Of Data- Series

A Two Dimensional Data Structure With Columns Of Potentially Different Types Of Data- Dataframe

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df.head()
```

```python
df.dropna(how='all', axis = 1)
```

```python
df.drop_duplicates(subset = ['name'])
```

```python
df.drop_duplicates(subset = ['name'], keep = 'last')
```

```python
df.drop_duplicates(subset = ['name'], keep = False)
```
